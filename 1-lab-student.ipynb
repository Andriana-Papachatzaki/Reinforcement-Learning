### 2D Cliff World

*(Taken from "Reinforcement Learning: An Introduction, 2nd Edition, Richard S. Sutton, Andrew G. Barto, 2015, MIT Press.")*



- S (blue) is the starting position
- G (green) is the goal
- Orange cells are cliff cells

At each timestep, the agent can be in only one cell. The agent has available the following actions:

1. Move Up
2. Move Down
3. Move Right
4. Move Left

If a move would make the agent go "out of bounds", then the move has no effect and the agent remains in place. The agent for each state-action pair $(s,a)$ receives the following rewards:

- $-100$ for an action that results in falling of the cliff
- $0$ for an action that results in stepping on goal
- $-1$ otherwise/for each step (even if the agents ends up in the same place)

The environment "stops" when the agent reaches the goal or steps falls off the cliff.

We first need to define the *Markov Decision Process* (MDP) for this problem/world.

```python
# Let's first import modules
import numpy as np # Linear Algebra
import matplotlib.pyplot as plt # Plotting
import copy
```

```python
#We first define the state space. We want to define each cell with a tuple (i,j), and we want to save all cells/states in a list called 'states' (x goes from 0 to 3 and y from 0 to 10).
states = [(x, y) for x in range(4) for y in range(11)]

### We now define the action space. This is a list of all possible actions (we use strings to represent them)
actions = ['Move Up', 'Move Down', 'Move Right', 'Move Left']

# Helper functions to check if a state is a cliff or terminal state (we need those since our MDP "stops" when arriving in one of those states)
def is_cliff(s):
    return (s[0] == 3 and s[1] > 0 and s[1] < 10)

def is_terminal(s):
    return (s == (3, 10))

### Define a world_step function (aka the system dynamics). The function takes as input the currest state (state) and the action to perform and returns the new state (a tuple).
def world_step(state, action):
    ### TODO: ADD YOUR CODE HERE
    row, col = state
    if(action == 'Move Up'):
        next_state = (row-1, col)
    elif(action == 'Move Right'):
        next_state = (row, col+1)
    elif(action == 'Move Down'):
        next_state = (row+1, col)
    elif(action == 'Move Left'):
        next_state = (row, col-1)
    if next_state[0] < 0 or next_state[0] >= 4 or next_state[1] < 0 or next_state[1] >= 11:
            next_state = (row, col)
    return next_state
    ### END of TODO

# Let's define the reward table/function (this is a python dictionary: we want to be able to query it with state-action pairs/tuples (state, action)).
reward = {}
## example indexing:
## s = states[0]
## a = actions[0]
## reward[(s, a)] = -1.
for s in states:
    for a in actions:
        ### Fill with correct value
        ### TODO: ADD YOUR CODE HERE
        next_state = world_step(s, a)
        if is_cliff(next_state):
            reward[(s, a)] = -100
        elif is_terminal(next_state):
            reward[(s, a)] = 0
        else:
            reward[(s, a)] = -1
### END of TODO
```

### SARSA: On-Policy TD Learning

1. Initialize $Q(\boldsymbol{s},\boldsymbol{a})=0, \forall\boldsymbol{s}\in\mathcal{S}, \forall\boldsymbol{a}\in\mathcal{A}$
2. At every step:
   1. we observe the state $\boldsymbol{s}$
   2. we select and execute an action according to our policy $\boldsymbol{a}\sim\pi(\boldsymbol{s})$
   3. we end up at state $\boldsymbol{s}'$ and receive reward $r$
   4. we select $\boldsymbol{a}'$ using $Q$ and $\boldsymbol{s}'$ (e.g. using $\epsilon-\text{greedy}$)
   5. $Q(\boldsymbol{s},\boldsymbol{a}) = Q(\boldsymbol{s},\boldsymbol{a}) + \alpha\Big(r + \gamma Q(\boldsymbol{s}',\boldsymbol{a}') - Q(\boldsymbol{s},\boldsymbol{a})\Big)$

Let's implement it.

```python
# First let's create the Q-function
N_states = len(states)
N_actions = len(actions)

# No need to keep an explicit policy representation, only the Q function is needed
Q = {}
### Initialize Q-function with zeros for all states. Q function is a dictionary that keys are tuples of the form (s,a)
def zero_Q():
    ### TODO: ADD YOUR CODE HERE
    for s in states:
        for a in actions:
            Q[(s, a)] = 0.0
    ### END of TODO
zero_Q()

# Policy given the Q-function
def pi(s, behavior="greedy", epsilon=0.1):
    ### Create the ε-greedy policy based on the Q function. The function should use the Q variable, and return an action. The argument "behavior" changes the behavior of the policy: if behavior="greedy", then the policy should act greedily, otherwise it should act ε-greedily.
    ### TODO: ADD YOUR CODE HERE
    if behavior == "greedy" or np.random.random() > epsilon:
        q_values = [Q[(s, a)] for a in actions]
        best_action = np.argmax(q_values)
        return actions[best_action]
    return np.random.choice(actions)
    ### END of TODO
```

```python
# Now we need to define a step_episode function that 1) receives a state, 2) selects an action using the policy, 3) steps the world forward, 4) computes the reward, 5) checks if the episode is done/finished, and 6) returns all the information
def step_episode(state, pi_behavior, pi_epsilon):
    info = {
        "state": copy.copy(state),
        "action": None,
        "next_state": None,
        "reward": None,
        "done": None
    }

    # We need to fill info with the correct values
    ### TODO: ADD YOUR CODE HERE
    action = pi(state, pi_behavior, pi_epsilon)
    next_state = world_step(state, action)
    current_state = next_state
    r = reward[(state, action)]
    if is_terminal(current_state):
        done = True
    elif is_cliff(current_state):
        done = False
    else:
        done = False

    info["action"] = action
    info["next_state"] = next_state
    info["reward"] = r
    info["done"] = done
    ### END of TODO
    return info

# Now we need to perform the SARSA update of the Q-function. We define a function that takes as input the information of the step_episode, and updates the Q-function dictionary (assume Q variable exists globally and update it in place)
def sarsa_update(info, alpha, gamma, pi_behavior, pi_epsilon):
    ### TODO: ADD YOUR CODE HERE
    state = info["state"]
    action = info["action"]
    next_state = info["next_state"]
    reward = info["reward"]
    done = info["done"]
    current_q = Q[(state, action)]
    
    if done:
        target = reward
    else:
        next_action = pi(next_state, pi_behavior, pi_epsilon)
        target = reward + gamma * Q[(next_state, next_action)]
    Q[(state, action)] = current_q + alpha * (target - current_q)
    ### END of TODO
```

```python
# Now we need to interact with the environment
N_episodes = 1000
gamma = 1. # episodic setting
alpha = 0.25
epsilon = 0.1
max_steps = 500

all_rewards = [] # for plotting
previous_reward = None
alpha_reward = 0.8

zero_Q() # zero out Q for new learning

for i in range(N_episodes):
    # We reset the world to the initial state
    state = (3, 0)

    # let's run one episode and collect the data
    rewards = []

    for _ in range(max_steps):
        # step world using the policy: we use ε-greedy while training/learning
        info = step_episode(state, "stochastic", epsilon)

        # Book-keeping
        rewards.append(info["reward"])

        # SARSA update
        sarsa_update(info, alpha, gamma, "stochastic", epsilon)

        if info["done"]:
            break
        state = copy.copy(info["next_state"])
    if previous_reward is None:
        previous_reward = np.sum(rewards)
    else:
        previous_reward = alpha_reward * previous_reward + (1. - alpha_reward) * np.sum(rewards)
    # print("Episode #" + str(i) + ": " + str(previous_reward))
    all_rewards.append(previous_reward)
```

```python
# Let's plot the learning process
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(range(len(all_rewards)), all_rewards, label='SARSA')

plt.show();
```

```python
# Now let's do an episode with the greedy policy (aka the best learned)
state = (3, 0) # we reset the world

# let's run one episode and collect the data
rewards = []

for _ in range(max_steps):
    # step world using the policy: we use ε-greedy while training/learning
    info = step_episode(state, "greedy", epsilon)

    # Book-keeping
    rewards.append(info["reward"])

    if info["done"]:
        break
    state = copy.copy(info["next_state"])
print('Episode reward is:', np.sum(rewards))
```

### Q-Learning: Off-Policy TD Learning

1. Initialize $Q(\boldsymbol{s},\boldsymbol{a})=0, \forall\boldsymbol{s}\in\mathcal{S}, \forall\boldsymbol{a}\in\mathcal{A}$
2. At every step:
   1. we observe the state $\boldsymbol{s}$
   2. we select and execute an action according to our policy $\boldsymbol{a}\sim\pi(\boldsymbol{s})$
   3. we end up at state $\boldsymbol{s}'$ and receive reward $r$
   4. $Q(\boldsymbol{s},\boldsymbol{a}) = Q(\boldsymbol{s},\boldsymbol{a}) + \alpha\Big(r + \gamma\max_zQ(\boldsymbol{s}',\boldsymbol{z}) - Q(\boldsymbol{s},\boldsymbol{a})\Big)$

Let's implement it.

```python
# First let's create the Q-function
N_states = len(states)
N_actions = len(actions)

# No need to keep an explicit policy representation, only the Q function is needed
Q = {}

zero_Q()
```

```python
# Now we need to perform the Q-learning update of the Q-function. We define a function that takes as input the information of the step_episode, and updates the Q-function dictionary (assume Q variable exists globally and update it in place)
def q_learning_update(info, alpha, gamma):
    ### TODO: ADD YOUR CODE HERE
    state = info["state"]
    action = info["action"]
    next_state = info["next_state"]
    r = info["reward"]
    done = info["done"]
    
    current_q = Q[(state, action)]
    
    if done:
        target = r
    else:
        max_next_q = max([Q[(next_state, a)] for a in actions])
        target = r + gamma * max_next_q
        
    Q[(state, action)] += alpha * (target - Q[(state, action)])
    ### END of TODO
```

### Expected SARSA: Off-Policy TD Learning

1. Initialize $Q(\boldsymbol{s},\boldsymbol{a})=0, \forall\boldsymbol{s}\in\mathcal{S}, \forall\boldsymbol{a}\in\mathcal{A}$
2. At every step:
   1. we observe the state $\boldsymbol{s}$
   2. we select and execute an action according to our policy $\boldsymbol{a}\sim\pi(\boldsymbol{s})$
   3. we end up at state $\boldsymbol{s}'$ and receive reward $r$
   4. $Q(\boldsymbol{s},\boldsymbol{a}) = Q(\boldsymbol{s},\boldsymbol{a}) + \alpha\Big(r + \gamma\sum_{a'}[\pi(\boldsymbol{a}'|\boldsymbol{s}')Q(\boldsymbol{s}',\boldsymbol{a}')] - Q(\boldsymbol{s},\boldsymbol{a})\Big)$

Let's implement it.

```python
# First let's create the Q-function -- just copy paste from previous
N_states = len(states)
N_actions = len(actions)

# No need to keep an explicit policy representation, only the Q function is needed
Q = {}

# Initialize Q-function with zeros for all states. Q function is a dictionary that keys are tuples of the form (s,a). Copy paste from above.
zero_Q()
```

```python
# Now we need to perform the expected SARSA update of the Q-function (we use the soft-max policy π', as in the slides). We define a function that takes as input the information of the step_episode, and updates the Q-function dictionary (assume Q variable exists globally and update it in place)
def exp_sarsa_update(info, alpha, gamma):
    ### TODO: ADD YOUR CODE HERE
    state = info["state"]
    action = info["action"]
    next_state = info["next_state"]
    r = info["reward"]
    done = info["done"]
    
    if done:
        target = r
    else:
        q_next = [Q[(next_state, a)] for a in actions]
        best_a_idx = np.argmax(q_next)
        n_a = len(actions)
        expected_q = 0
        for i in range(n_a):
            prob = epsilon / n_a
            if i == best_a_idx:
                prob += (1.0 - epsilon)
            expected_q += prob * q_next[i]
            
        target = r + gamma * expected_q
        
    Q[(state, action)] += alpha * (target - Q[(state, action)])
    ### END of TODO
```
